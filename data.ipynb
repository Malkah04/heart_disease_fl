```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
df =df.sample(frac=1, random_state=42).reset_index(drop=True)   


h1, h2, t = np.split(df, [int(0.45 * len(df)), int(0.90 * len(df))])

h1.to_csv('heart_h1.csv', index=False)
h2.to_csv('heart_h2.csv', index=False)
t.to_csv('heart_t.csv', index=False)
```
